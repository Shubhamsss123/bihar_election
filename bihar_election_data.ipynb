## since we have large number of parties and AC numbers one hot encoding will result in very sparse data and using target encoding

```python
df.columns
```

```python
df1=df.drop(columns=['Votes','Valid Votes','Electors','Voteshare','Margin','Normlaized_Valid_Votes','Rank'])
df1
```

```python
X=df1[['AC No', 'Party', 'Alliance']]
y=df1[['Normalized_Votes']]
encoder = TargetEncoder(cols=['AC No', 'Party','Alliance'])
X_encoded = encoder.fit_transform(X, y)
X_encoded
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_encoded,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = model.predict(X_test)
r2_score(y_test, y_pred)
```

```python
r2_score(y_train, model.predict(X_train))
```

## catboost regressor:

```python
cat_features=['AC No', 'Party','Alliance']
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, test_size=0.3, random_state=42)
model1 = CatBoostRegressor(cat_features=cat_features, iterations=500, learning_rate=0.1, depth=6, verbose=0)
model1.fit(X_train1, y_train1)
```

```python
r2_score(y_train1, model1.predict(X_train1))
```

```python
r2_score(y_test1, model1.predict(X_test1))
```

```python
def predict_normalized_votes(ac_no, party, Alliance):
    # Convert inputs to a DataFrame format similar to training data
    input_data = pd.DataFrame({
        'AC No': [ac_no],
        'Party': [party],
        'Alliance': [Alliance],
    })

    # Predict using the trained model
    predicted_normalized_votes = model1.predict(input_data)
    return predicted_normalized_votes[0]
```

```python
future_ac_no = 1
future_party = 'JD(U)'
Alliance='NDA'
factor=df[(df['AC No']==future_ac_no)&(df['Year']==2020)]['Valid Votes'].values[0]*(1.1)
predicted_vote = predict_normalized_votes(future_ac_no, future_party, Alliance)
print(f"Predicted Normalized Votes for future data: {round(predicted_vote*factor)}")
```
